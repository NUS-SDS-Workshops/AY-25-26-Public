### 0. Setting up MLClient

```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential

ml_client = MLClient.from_config(
        credential=DefaultAzureCredential(),
        file_name="config.json", 
        # default is config.json, only specify if you change your file name :) 
        )
```

### 1. Upload the Data as Asset

**IMPORTANT**: By default, data asset will be create in the default blob storage, i.e., `workspaceblobstorage`. Thus, we don't need to create one ourselves. 

**NOTE**: If you have `azcopy` installed in your device, you can also use the following command to upload the data, which is actually very much faster.

`azcopy cp '/path/to/your/fmnist/folder' 'https://<storage-account-name>.blob.core.windows.net/<container>/<SAS-token>' --recursive=true`

To get the SAS-token, navigate to your Azure Portal > Storage Account > Container > ⋮ > Generate SAS > Add necessary permissions > Generate and copy the key

But in the following code, we **WILL NOT** be using `azcopy` in consideration of the general audience.

```python
from pathlib import Path

# Update your local root accordingly if necessary
local_root = Path("../data")
```

```python
import os 
from azure.ai.ml.entities import Data
from azure.ai.ml.constants import AssetTypes

train_v1_data = Data(
    name="train_v1",
    version="1",
    description="Fashion MNIST 40k train data. Contains 10 folders that correspond to 10 classes",
    path=os.path.join(local_root, "train_v1"),
    type=AssetTypes.URI_FOLDER,
)

train_v1 = ml_client.data.create_or_update(train_v1_data)
```

```python
train_v1 = ml_client.data.get(name='train_v1', version=1)
print(f"Data asset created. Name: {train_v1.name}, version: {train_v1.version}")
```

```python
train_v2_data = Data(
    name="train_v2",
    version="1",
    description="Fashion MNIST 20k train data. Contains 10 folders that correspond to 10 classes",
    path=os.path.join(local_root, "train_v2"),
    type=AssetTypes.URI_FOLDER,
)

train_v2 = ml_client.data.create_or_update(train_v2_data)
```

```python
train_v2 = ml_client.data.get(name='train_v2', version=1)
print(f"Data asset created. Name: {train_v2.name}, version: {train_v2.version}")
```

```python
test_data = Data(
    name="test",
    version="1",
    description="Fashion MNIST 10k test data. Contains 10 folders that correspond to 10 classes",
    path=os.path.join(local_root, "test"),
    type=AssetTypes.URI_FOLDER,
)

test = ml_client.data.create_or_update(test_data)
```

```python
test = ml_client.data.get(name='test', version=1)
print(f"Data asset created. Name: {test.name}, version: {test.version}")
```
